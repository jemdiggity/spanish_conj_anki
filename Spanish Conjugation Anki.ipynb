```python
import requests
from bs4 import BeautifulSoup
from bs4 import SoupStrainer
import sys
import csv
```

```python
# Import a list of verbs. It is these words that we will look up
# the conjugations for.
verb_list = []
with open('verb_top100.txt', newline='') as csvfile:
    reader = csv.reader(csvfile, delimiter=' ', quotechar='|')
    for row in reader:
        verb_list.append(row[0])
```

```python
# For each verb, grab the conjugation from wordreference.com and
# make a dictionary of verb->mood->tense->conjugation.
big_dict = {}
for verb in verb_list:
    big_dict[verb] = {}
    r = requests.get('https://www.wordreference.com/conj/ESverbs.aspx?v=' + verb)
    soup = BeautifulSoup(r.content, 'html.parser')
    for moods in soup.find_all('div', "aa"):
        mood = str(moods.h4.string)
        if mood not in big_dict[verb]:
            big_dict[verb][mood] = {}
        for tense in moods.find_all('table'):
            print(tense.find_all('tr')[0].get_text().split('ⓘ')[0])
# For some tenses, the tag contains not just the name but alternate names. Parse this out.
            tense_name = tense.find_all('tr')[0].get_text().split('ⓘ')[0]
            big_dict[verb][mood][tense_name] = {}
            conjugation = []
            for person in tense.find_all('td'):
                conjugation.append(str(person.get_text()))
            print(conjugation)
            big_dict[verb][mood][tense_name] = conjugation
```

```python
# Create a csv file with "Verb (Tense)" in column 1 and it's conjugated forms in column 2.
# Column 3 has the tense again so it can be imported as a tag in Anki.
# Import into Anki with tab separation.
with open('dictionary.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile, delimiter='\t', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    for verb, moods in big_dict.items():
        for mood, tenses in moods.items():
            for tense, persons in tenses.items():
#                 print(verb, mood, tense, persons)
                pronouns = ['yo','tú','él, ella, Ud.','nosotros','vosotros','ellos, ellas, Uds.','vos' ]
                conjugation = []
                for item in zip(pronouns, persons):
                    conjugation.append(' '.join(item))
                tense_pretty = '{} {}'.format(mood, tense)
                if 'Indicativo' in mood or 'Formas compuestas comunes' in mood:
                    tense_pretty = tense
                if 'Tiempos compuestos del subjuntivo' in mood:
                    tense_pretty = "subjuntivo {}".format(tense)
                tense_pretty = tense_pretty.lower()
                print("{} ({})".format(verb, tense_pretty))
                writer.writerow(["{} ({})".format(verb, tense_pretty), '<br>'.join(conjugation), "{}".format(tense_pretty.replace(' ', '_'))])
```
